# Small examples for using classes and objects

```python
class FlightTicket:
    def __init__(self, name, ticket_type, price, baggage_allowance):
        self.name = name
        self.ticket_type = ticket_type
        self.price = price
        self.baggage_allowance = baggage_allowance

    def display_details(self):
        print(f"Name: {self.name}")
        print(f"Ticket Type: {self.ticket_type}")
        print(f"Price: ${self.price}")
        print(f"Baggage Allowance: {self.baggage_allowance} kg")
        print("-" * 30)

    def upgrade(self, extra_price, extra_baggage):
        self.price += extra_price
        self.baggage_allowance += extra_baggage
        print(f"{self.name}'s {self.ticket_type} ticket upgraded! New price: ${self.price}, New baggage: {self.baggage_allowance} kg")

alice = FlightTicket("Alice", "Business", 1200, 40)
bob = FlightTicket("Bob", "Premium Economy", 800, 30)
charlie = FlightTicket("Charlie", "Economy", 400, 20)
dhl_cargo = FlightTicket("DHL Logistics", "Cargo", 200, 100)

alice.display_details()
bob.display_details()
charlie.display_details()
dhl_cargo.display_details()

print("Upgrading Charlie's ticket...\n")
charlie.upgrade(100, 5)

print("\nAfter upgrade:")
charlie.display_details()
alice.display_details()  

# First try to predict what going to happen and then run the code
```

```python
class FlightTicket:
    def __init__(self, name, ticket_type, price, baggage_allowance):
        self.name = name
        self.ticket_type = ticket_type
        self.price = price
        self.baggage_allowance = baggage_allowance

    def display_details(self):
        print(f"Name: {self.name}")
        print(f"Ticket Type: {self.ticket_type}")
        print(f"Price: ${self.price}")
        print(f"Baggage Allowance: {self.baggage_allowance} kg")
        print("-" * 30)

    def upgrade(self, extra_price, extra_baggage=None):
        self.price += extra_price
        if extra_baggage is not None:
            self.baggage_allowance += extra_baggage
            print(f"{self.name}'s {self.ticket_type} ticket upgraded! New price: ${self.price}, New baggage: {self.baggage_allowance} kg")
        else:
            print(f"{self.name}'s {self.ticket_type} ticket upgraded! New price: ${self.price}")


alice = FlightTicket("Alice", "Business", 1200, 40)
charlie = FlightTicket("Charlie", "Economy", 400, 20)


alice.upgrade(200, 10)  
charlie.upgrade(50)     

alice.display_details()
charlie.display_details()

#  This is a classic example of function overloading. May be helpful
```

```python
class Ticket:
    def __init__(self, name, ticket_type, price):
        self.name = name
        self.ticket_type = ticket_type
        self.price = price

    def display_details(self):
        print(f"Name: {self.name}")
        print(f"Ticket Type: {self.ticket_type}")
        print(f"Price: ${self.price}")

# A child class for flight ticket
class FlightTicket(Ticket):
    def __init__(self, name, ticket_type, price, baggage_allowance):
        super().__init__(name, ticket_type, price)
        self.baggage_allowance = baggage_allowance
        self.mode = "Flight"

    def display_details(self):
        print(f"Mode : {self.mode}")
        super().display_details()
        print(f"Baggage Allowance: {self.baggage_allowance} kg")
        print("-" * 30)

# A child class for rail ticket
class RailTicket(Ticket):
    def __init__(self, name, ticket_type, price, seat_number):
        super().__init__(name, ticket_type, price)
        self.seat_number = seat_number
        self.mode = "Rail"

    def display_details(self):
        print(f"Mode : {self.mode}")
        super().display_details()
        print(f"Seat Number: {self.seat_number}")
        print("-" * 30)

flight = FlightTicket("Alice", "Business", 1200, 40)
rail = RailTicket("Bob", "AC First Class", 500, "A1-23")

flight.display_details()
rail.display_details()

# Classic example of Inheritance
```
